<h1>Pre-processing</h1>

```python
#Removing White space in columns
for column in df:
    df.columns = df.columns.str.replace(' ','_')
df.drop('Unnamed:_0', axis=1, inplace=True)
df.columns
```

<h1>Visualization</h1>

```python
bin_labels = ['Okay','Good','Great']
plt.gcf().set_size_inches(23, 10)

plt.subplot(2,4,1)
Well_versed_with_the_subject_cat = 0
Well_versed_with_the_subject_cat = pd.qcut(df.Well_versed_with_the_subject,labels=bin_labels,q=3)
ax = sns.countplot(Well_versed_with_the_subject_cat)
ax.bar_label(ax.containers[0])

plt.subplot(2,4,2)
Explains_concepts_in_an_understandable_way_cat = pd.qcut(df.Explains_concepts_in_an_understandable_way,labels=bin_labels,q=3)
ax = sns.countplot(Explains_concepts_in_an_understandable_way_cat)
ax.bar_label(ax.containers[0])

plt.subplot(2,4,3)
Use_of_presentations_cat = pd.qcut(df.Use_of_presentations,labels=bin_labels,q=3)
ax = sns.countplot(Use_of_presentations_cat)
ax.bar_label(ax.containers[0])

plt.subplot(2,4,4)
Degree_of_difficulty_of_assignments_cat = pd.qcut(df.Degree_of_difficulty_of_assignments,labels=bin_labels,q=3)
ax = sns.countplot(Degree_of_difficulty_of_assignments_cat)
ax.bar_label(ax.containers[0])

plt.subplot(2,4,5)
Solves_doubts_willingly_cat = pd.qcut(df.Solves_doubts_willingly,labels=bin_labels,q=3)
ax = sns.countplot(Solves_doubts_willingly_cat)
ax.bar_label(ax.containers[0])

plt.subplot(2,4,6)
Structuring_of_the_course_cat = pd.qcut(df.Structuring_of_the_course,labels=bin_labels,q=3)
ax = sns.countplot(Structuring_of_the_course_cat)
ax.bar_label(ax.containers[0])

plt.subplot(2,4,7)
Provides_support_for_students_going_above_and_beyond_cat = pd.qcut(df.Provides_support_for_students_going_above_and_beyond,labels=bin_labels,q=3)
ax = sns.countplot(Provides_support_for_students_going_above_and_beyond_cat)
ax.bar_label(ax.containers[0])

plt.subplot(2,4,8)
Course_recommendation_based_on_relevance_cat = pd.qcut(df.Course_recommendation_based_on_relevance,labels=bin_labels,q=3)
ax = sns.countplot(Course_recommendation_based_on_relevance_cat)
ax.bar_label(ax.containers[0])
```

```python
#Creating New DataFrame with Categorical Values
new_df = df.filter(['Student_ID'], axis=1)

new_df['Well_versed_with_the_subject_cat'] = Well_versed_with_the_subject_cat
new_df['Explains_concepts_in_an_understandable_way_cat'] = Explains_concepts_in_an_understandable_way_cat
new_df['Use_of_presentations_cat'] = Use_of_presentations_cat
new_df['Degree_of_difficulty_of_assignments_cat'] = Degree_of_difficulty_of_assignments_cat
new_df['Solves_doubts_willingly_cat'] = Solves_doubts_willingly_cat
new_df['Structuring_of_the_course_cat'] =  Structuring_of_the_course_cat
new_df['Provides_support_for_students_going_above_and_beyond_cat'] =  Provides_support_for_students_going_above_and_beyond_cat
new_df['Course_recommendation_based_on_relevance_cat'] =  Course_recommendation_based_on_relevance_cat

new_df
```
